```python
import numpy as np
def findyear(x) :
    for i in x.split():
        if i.isdigit() and len(i) == 4:
            return int(i)
    return np.nan

wine['year'] = wine['title'].apply(findyear)
wine.year.unique()
```

Sommige wijnerijen hebben in hun naam specifiek een jaartal. Dit is een groot getal wat niet laat zien wanneer het is geproduceerd. Ik wilde een lijst met wanneer specifieke wijnen zijn geproduceerd. 

Het is verstandig om nog een conditie toe te voegen. Getallen die kleiner of groter zijn dan een bepaalde range houd die er dan uit.

```python
import numpy as np
def findyear(x) :
    for i in x.split():
        if i.isdigit() and len(i) == 4 and int(i) < 2025 and int(i) > 1950:
            return int(i)
    return np.nan

wine['year'] = wine['title'].apply(findyear)
wine.year.unique()
```

np.nan betekent als je geen jaartal hebt gevonden geef dan de terug als nan. Not a Number. Omdat we nan gebruiken worden de andere getallen geschreven als float.
