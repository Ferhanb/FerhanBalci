```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
```

```python
np.random.seed(42)

x = np.arange(50,150,1)
y = x*5+ np.random.normal(0,20,100)

plt.plot(x,y,'.')
plt.xlabel('Oppervlak in vierkante meter')
plt.ylabel('Prijs in duizend euro')
plt.title('Huizenprijzen')
plt.show()
```

Je krijgt bij random.normal voeg bij elk datapunt van de y-waarde een getal tussen 0 en 20 op. Doe dit voor 100 datapunten.  Je had bijvoorbeeld het getal 150 plak daar random getal bij op.

Voor random.seed kan je een random getal toevoegen. Dan categoriseert het model het elke keer random op dezelfde manier.

```python
np.random.seed(42)

x = np.arange(50,150,1)
y = x*5+ np.random.normal(0,20,100)

plt.plot(x,y,'.')
plt.xlabel('Oppervlak in vierkante meter')
plt.ylabel('Prijs in duizend euro')
plt.title('Huizenprijzen')


plt.xlim(45,150)
plt.ylim(0,800)

plt.show()
```

De xlim en ylim kapt een gedeelte van de grafiek af.

```python
x
```

```python
x = x.reshape(100,1)
y = y.reshape(100,1)
```

reshape maakt van het model een kolom. 1 dimensionaal zie je alle getallen onder elkaar. Als je (100,2) zet dan kan je 2 kolommen krijgen.

```python
x
```

```python
linear = linear_model.LinearRegression()
result = linear.fit(x,y)
```

```python
result.coef_
```

Dit is a. Die komt terug van  y= 5x +np. etcetera

```python
result.intercept_
```

INTERCEPT is b

```python
predict_y = result.predict([[200]])
predict_y
```

Met deze formule laat je het model uitrekenen hoeveel een 200 m2 huis kost.

```python
result.score(x,y)
```

Score geeft de r2 (r kwadraat) eruit. Dit is een mate hoe goed het model al deze punten fit. Tussen 0 en 1. 0 is geen punten. 1 is perfect.

```python
model = result.predict(x)
```

```python


plt.plot(x,y,'.')
plt.plot(x,model, '.', color="red")
plt.xlabel('Oppervlak in vierkante meter')
plt.ylabel('Prijs in duizend euro')
plt.title('Huizenprijzen')


plt.xlim(45,150)
plt.ylim(0,800)

plt.show()
```

```python
residuals = model - y
plt.hist(residuals)
plt.show()
```

Het gene wat overblijft Lineaire Regressie - residuals
